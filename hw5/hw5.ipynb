```python
# Смотрим какие типы данных в сете
# Все поля числовые, кроме диагноза, присутствует поле которое не заполнено вообще
# В остальных полях нет пропущенных значений
data.info()
```

```python
# Удалим не нужное поле Unnamed: 32
data.drop(columns={'Unnamed: 32'},inplace=True)
```

```python
# Строим гистограммы с группировкой и раскраской по искомой переменной
for el in data.columns:
    plt.figure()
    sns.histplot(data=data, hue='diagnosis', x=el)
```

```python
# Конвертируем диагноз в числовой вид
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
data['diagnosis'] = lbl.fit_transform(data['diagnosis'])
```

```python
# Строим heatmap для всех переменных сета
plt.figure(figsize=(20,16))
sns.set(font_scale=1.25)
sns.heatmap(data.corr(), cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 12},\
                 vmin=-1, center=0, cmap=sns.color_palette('coolwarm',1000))
plt.show()
```

```python
# Видно что есть поля сильно скоррелированные между собой.
# Найдем поля, которые сильно коррелируют.
corr_list = []
correlation = data.corr()
for i in range(len(correlation)):
    for j in range(i):
        if correlation.iloc[i,j] > 0.90 :
            corr_list.append([correlation.columns[i], correlation.columns[j]])
corr_list
```

```python
# Построим scatterplot для полей которые сильно скоррелированы
# На всех графиках видна линейная зависимоть
for el in corr_list:
    plt.figure()
    sns.scatterplot(data=data, x=el[0], y=el[1])
```

```python
# Удаляем поля, которые сильно скоррелированы
for el in corr_list:
    if el[1] in data.columns:
        data.drop(columns=el[1], inplace=True)
data.shape
```

```python
# Нарисуем boxplot для оставшихся полей
for el in data.columns:
    if el != 'diagnosis':
        plt.figure()
        sns.boxplot(data=data, x='diagnosis', y=el)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score

def print_metrics(right_vals, pred_vals):
    acc = accuracy_score(right_vals, pred_vals)
    pre = precision_score(right_vals, pred_vals, average='weighted')
    recall = recall_score(right_vals, pred_vals, average='weighted')
    f1 = f1_score(right_vals, pred_vals, average='weighted')
    print('accuracy:  ', acc)
    print('precision: ', pre)
    print('recall:    ', recall)
    print('f1:        ', f1)
```

```python
# Разбиение сета на трейн и тест

X_train, X_test, y_train, y_test = train_test_split(data.drop(columns={'diagnosis'}), data['diagnosis'], test_size=0.3, random_state=36)
```

```python
# Приведение данных к одному масштабу

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# Строим можель с параметрами по умолчанию
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred_test = knn.predict(X_test_scaled)
print_metrics(y_test.values, pred_test)
```

```python
# ПРоведет проверку парметра числа соседей на кросс валидации

neighbors = range(1, 50)
accuracy_train = []
accuracy_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    accuracy_train.append(accuracy_score(knn.predict(X_train_scaled), y_train))
    accuracy_test.append(accuracy_score(knn.predict(X_test_scaled), y_test))
    
plt.plot(neighbors, accuracy_train, color='blue', label='train')
plt.plot(neighbors, accuracy_test, color='red', label='test')
plt.title("Max test quality: {:.3f}\nBest k: {}".format(max(accuracy_test), np.argmax(accuracy_test)+1))
plt.legend()
plt.show()
```

```python
# Настроим классификатор с учетом оптимального числа соседей, которое было получено на кросс валидации.
# Модель стала работать немного лучше

knn = KNeighborsClassifier(n_neighbors=14)
knn.fit(X_train_scaled, y_train)
pred_test = knn.predict(X_test_scaled)
print_metrics(y_test.values, pred_test)
```
